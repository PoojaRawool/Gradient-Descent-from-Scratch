```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta= -1                               
learning_rate = 0.09                           
convergencelimit = 0.0000000000001         #This tells us when to stop the algorithm
previous_step_size = 1                
min_change_in_cost_fn = 0.00000001    #minimum change in cost function to proceed further
delta_cost_fn_for_step = 10000000000  #actual change in cost function for current step
max_iterations = 10000                # maximum number of iterations
itercount = 0                         #iteration counter
fn = lambda beta: ((beta+2)*(beta+2)) - 5    #The cost function
df = lambda beta: 2*(beta+2)                #Cost function Gradient 
cost_fun_val = []                     #Empty list to store successive cost function value
beta_val = []                         #Empty list to store successive beta value

begintime = time.time()
costfn = fn(beta)
#Iterate and recompute bk till you reach maxiterations or the desired precision in stepsize

while previous_step_size > convergencelimit and itercount < max_iterations and delta_cost_fn_for_step > min_change_in_cost_fn and costfn != 0: 
    beta_val.append(beta)                       #Store current beta value in array
    cost_fun_val.append(costfn)               #Store current cost function value in array
    
    prev_beta = beta                             #Store current beta value in prev_beta
    prev_costfn = costfn
    beta = beta - learning_rate * df(prev_beta)               #Compute the next value of beta using gradient and learning_rate
    
    #We will stop if the step we take is too small. So we compute step size
    previous_step_size = abs(beta - prev_beta)    #measure the magnitude of step size to compare with precision
    costfn = fn(beta)
    delta_cost_fn_for_step = abs(costfn - prev_costfn)
    
    itercount += 1                            #increment iteration count to compare with maxiterations
    print("Iteration",itercount,"\nX value is",beta) #Print iterations
    
print("The local minimum occurs at {} and is equal to {}".format(beta, costfn))
endtime = time.time()
print(f"Program execution time= {endtime - begintime} seconds") 
```

```python
plt.figure(figsize=(8,7))
plt.scatter(beta_val,cost_fun_val,color="blue")
plt.show()
```

```python
plt.figure(figsize=(8,7))
plt.scatter(beta_val,cost_fun_val,color="blue")
plt.xlim([-2.2,-1.4])
plt.show()
```
